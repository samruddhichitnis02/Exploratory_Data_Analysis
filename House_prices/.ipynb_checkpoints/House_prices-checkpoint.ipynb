# Checking for Null Values

```python
fig, ax = plt.subplots(figsize = (10, 6))
sb.heatmap(data.isna(),
           cbar = False,
           yticklabels = False,
           cmap = 'Blues')
plt.show()
```

removing  Id and the features with 30% or less data.

```python
len(data['Id'].unique())
```

```python
data.drop(['Id'], axis = 1, inplace = True)
```

```python
dataset = data[[column for column in data if data[column].count() / len(data) >= 0.3]]
```

```python
dataset.sample(2)
```

```python
print("List of dropped columns:", end=" ")
for c in data.columns:
    if c not in dataset.columns:
        print(c, end=", ")
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
sb.heatmap(dataset.isna(),
           cbar = False,
           yticklabels = False,
           cmap = 'Blues')
plt.show()
```

# Finding the Correlation between the features and label

```python
data_corr = dataset.corr()['SalePrice'][:-1] 
correlation = data_corr[abs(data_corr) >= 0.5].sort_values(ascending = False)
```

```python
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(correlation), correlation))
```

# Feature to feature relationship

```python
# We already examined SalePrice correlations
corr = dataset.drop('SalePrice', axis = 1).corr()
plt.figure(figsize=(15, 10))

sb.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
           cmap = '',
           vmax = 1.0,
           vmin = -1.0,
           linewidths = 0.1,
           annot = True,
           annot_kws = {"size": 10},
           square = True);
```

```python
quantitative_features_list = ['LotFrontage', 'LotArea', 'MasVnrArea', 
                              'BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', 
                              '1stFlrSF','2ndFlrSF', 'LowQualFinSF', 
                              'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 
                              'FullBath', 'HalfBath','BedroomAbvGr', 
                              'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
                              'GarageCars', 'GarageArea', 'WoodDeckSF', 
                              'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 
                              'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice']
df_quantitative_values = data[quantitative_features_list]
df_quantitative_values.head()
```

```python
features_to_analyse = [x for x in quantitative_features_list if x in correlation]
features_to_analyse.append('SalePrice')
```

Distribution of quantitative features which have good correlation with SalesPrice

```python
fig, ax = plt.subplots(2, 4, figsize = (18, 12))
for i, ax in enumerate(fig.axes):
    if i< len(features_to_analyse) -1: #-1 since it also has label('SalePrice')
        sb.regplot(x = features_to_analyse[i],
                   y = 'SalePrice',
                   data = dataset[features_to_analyse],
                   ax = ax)
```

We can see that features such as TotalBsmtSF, 1stFlrSF, GrLivArea have a big spread
